## What is DataFrame?  
A Dataframe is a 2D data structure that stores different types of data (i.e. integer, float, string, boolean) in rows and columns (i.e. Rectangular data)  
Where,  
Rows --> observations (records, entities)  
Columns --> features (variables, attributes)  

## Creating DataFrame:

### 1. From Dictionary

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 22],
    'City': ['New York', 'London', 'Paris']
}
df = pd.DataFrame(data)
print(df)
```

### 2. From List of List

```python
data = [
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'London'],
    ['Charlie', 22, 'Paris']
]
df = pd.DataFrame(data, columns=['Name', 'Age', 'City'])
print(df)
```

### 3. From List of Dictionaries

### Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### BMI Distribution 

```python
plt.hist(bmi_df["BMI"], bins=25, color="skyblue", edgecolor="black")
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("frequency")
plt.show()
```

#### BMI Categories Distribution

```python
# define BMI categories
def categorize(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal"
    elif 25 <= bmi < 30:
        return "Overweight"
    elif 30 <= bmi < 35:
        return "Obese Class I"
    else:
        return "Obese Class II"
```

```python
bmi_df["Categories"] = bmi_df["BMI"].apply(categorize)
```

```python
bmi_df
```

```python
category_order = ['Underweight', 'Normal', 'Overweight', 'Obese Class I', 'Obese Class II']
```

```python
# count number of peoples in each categories
bmi_counts = bmi_df["Categories"].value_counts().reindex(category_order)
```

```python
bmi_counts
```

```python
plt.bar(bmi_counts.index, bmi_counts.values, color=["skyblue", "lightgreen", "orange", "red"])
plt.xlabel("BMI Categories")
plt.ylabel("Number of peoples")
plt.title("BMI Categories Distribution")
plt.xticks(rotation=45)
plt.show()
```

### BMI Categories Distribution by Gender 

```python
bmi_gender_counts = bmi_df.groupby(['Categories', 'Gender']).size().unstack(fill_value=0)

categories = bmi_gender_counts.index
male_counts = bmi_gender_counts['Male']
female_counts = bmi_gender_counts['Female']
```

```python
bmi_gender_counts
```

```python
categories
```

```python
male_counts
```

```python
x = np.arange(len(categories))
width = 0.35

plt.bar(x-width/2, male_counts, width, label='Male', color='steelblue')
plt.bar(x+width/2, female_counts, width, label='Female', color='lightcoral')

plt.xlabel('BMI Categories')
plt.ylabel('Number of peoples')
plt.title('BMI Categories Distribution by Gender')
plt.xticks(x, categories, rotation=45)
plt.legend()
plt.show()
```
