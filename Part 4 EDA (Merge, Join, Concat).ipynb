## Merging:
In pandas, merging is the process of combining/joining two or more Dataframes or Series based on common columns.  
It is similar to SQL database join

### Forms of joins:

1. inner join
2. outer join
3. left join
4. right join

```python
# Creating dataFrames

# DataFrame 1 contains persons names with ids and ages
df1 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Charlie", "Diana"],
    "age": [25, 30, 35, 40]
})

# DataFrame 2 contains persons IDs living in the city with income
df2 = pd.DataFrame({
    "id": [3, 4, 5, 6],
    "city": ["London", "Paris", "Berlin", "Rome"],
    "income": [70000, 80000, 50000, 60000]
})

df1, df2
```

#### 1. Inner Join:  (Default)
Return only rows with matching keys in both DataFrames

```python
inner_join = pd.merge(df1, df2, on='id')
```

```python
inner_join
```

#### 2. Left Join:  
Combines two dataframes based on common keys, preserving all rows from the left dataframe and including matching rwos from the right dataframe

```python
left_join = pd.merge(df1, df2, on='id', how='left')
```

```python
left_join
```

#### 3. Right Join:
Combines two dataFrames based on common keys, preserving all rows from the right dataframe and including matching rows from the left dataframe

```python
right_join = pd.merge(df1, df2, on='id', how='right')
```

```python
right_join
```

#### 4. Outer Join:
Combines all rows of both dataframes

```python
outer_join = pd.merge(df1, df2, on='id', how='outer')
```

```python
outer_join
```

### Merge on columns with different names

```python
df1 = df1.rename(columns={'id': 'left_id'})
df2 = df2.rename(columns={'id': 'right_id'})
```

```python
df1, df2
```

```python
outer_join = pd.merge(df1, df2, left_on='left_id', right_on='right_id', how='outer')
```

```python
outer_join
```

### Merging using multiple keys

```python
df3 = pd.DataFrame({
    "year":[2020,2020,2021],
    "country":["US","UK","US"],
    "val":[10,20,30]
})

df4 = pd.DataFrame({
    "year":[2020,2021],
    "country":["US","US"],
    "extra":[100,200]
})
```

```python
left_join = pd.merge(df3, df4, on=['year', 'country'], how='outer')
```

```python
left_join
```

## Join:  
In pandas, .join() is used to combine columns of two dataframes base on their indexes

```python
df1 = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35]
}, index=["a", "b", "c"])

df2 = pd.DataFrame({
    "city": ["London", "Paris", "Berlin"],
    "salary": [70000, 80000, 65000]
}, index=["a", "b", "d"])
```

```python
# defalut in left join
df1.join(df2)
```

```python
df1.join(df2, how='inner')
```

```python
df1.join(df2, how='outer')
```

```python
df1.join(df2, how='right')
```

#### Join on specific column (instead of index)

```python
left = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"]
})

right = pd.DataFrame({
    "id": [1, 2, 4],
    "city": ["London", "Paris", "Berlin"]
})


left.join(right.set_index('id'), on='id')
```

## concat:  
it is used to stack or combine multiple DataFrames or Series along a particular axis (rows or columns).

```python
df1 = pd.DataFrame({
    "A": ["A0", "A1", "A2"],
    "B": ["B0", "B1", "B2"]
})

df2 = pd.DataFrame({
    "A": ["A3", "A4", "A5"],
    "B": ["B3", "B4", "B5"]
})
```

```python
# vertical concatenation --> axis=0  (Default)
pd.concat([df1, df2])
```

```python
# horizontal concatenation --> axis=1
pd.concat([df1, df2], axis=1)
```

#### Concatenating with different columns

```python
df3 = pd.DataFrame({
    "A": ["A0", "A7"],
    "C": ["C6", "C7"]
})
```

```python
pd.concat([df1, df3])
```
