# Data Cleaning:
Data cleaning in the process of identifying and correcting (or removuing) errors, missing values in your dataset so that your analysis is reliable

Data cleaning workflow:

This workflow covers almost every common cleaning scenario:

a. Missing data  
b. Duplicates  
c. Special characters  
d. Invalid values  
e. Inconsistent text  
f. Correct data types  

```python
import pandas as pd
import numpy as np
```

```python
# creating dummy data (with special charactors, missing values, duplicates, and invalid data)

data = {
    'Name': ['Alice ', '@Bob', 'Charlie$', 'David#', 'Eve%', 'Frank*', 'Alice ', np.nan],
    'Age': [25, np.nan, 35, 45, 28, -5, 25, 30],
    'Gender': ['Female', 'Male', 'Male', np.nan, 'Female', 'Male', 'Female', 'male'],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago', np.nan, 'Chicago', 'New York', 'Los Angeles'],
    'Salary': [50000, 60000, np.nan, 80000, 56000, 70000, 50000, 60000],
    'Comment': ['Good job!', 'Needs improvement  ', 'Paid $100', '50% discount', 'Excellent!!', 'Check* this', 'Good job!', '😊 Excellent']
}
```

```python
df = pd.DataFrame(data)
df
```

## Step 1: inspect the data

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

## Step 2: drop duplicates

removes exact duplicate rows

```python
df
```

```python
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df
```

## Handling missing Values

Drop rows with missing critical values

```python
df = df.dropna(subset=['Name', 'Gender'])   # by default axis = 0  (rows)
```

```python
df
```

df = df.dropna(axis=1)    # drop columns having NaN valuses

Fill missing values

Note: It is not feasible to used .loc for filling missin values, because in this we have to provide specific row and column labels. 
Instead of this we automate the process using fillna method. 
1. It find NaN value in the column 
2. Replace the value with the specified value

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df
```

```python
df['City'] = df['City'].fillna('unknown')
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
```

```python
df
```

## Clean text data
### Inconsistent data

```python
# remove leading/trailing spaces
df['Name'] = df['Name'].str.strip()
```

```python
# remove special characters
df['Name'] = df['Name'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
```

```python
df
```

```python
# Capitalize
df['Gender'] = df['Gender'].str.capitalize()  
```

```python
df
```

```python
# proper case for cities
df['City'] = df['City'].str.title()
```

```python
df
```

## Handling invalid/inconsistent data

```python
# filter rows with positive ages
df = df[df['Age'] > 0]
```

```python
df
```

### Converting Data Types

```python
df.dtypes
```

```python
# convert data types
df['Age'] = df['Age'].astype(int)
#df['Salary'] = df['Salary'].astype(int)
```

```python
df.dtypes
```

```python
df
```

## Rename columns for consistency

```python
# Rename columns for consistency
df.rename(columns={'Name': 'Employee_Name', 'Comment': 'Performance'}, inplace=True)
```

```python
df
```

## Final cleaned data

```python
df
```

### dealing with data that might contain non-numeric values (e.g., strings) in a column that should ideally be numerical.

```python
ddf = pd.DataFrame({
    'Age': ['22', '44', 'twenty']
})
```

In this DataFrame the Column Age should be numeric but it is provide in strings.  
To convert string to numerical value and handling invalid number string such as 'twenty' 

```python
# coerce will handling invalid string and replace it with NaN
ddf = pd.to_numeric(ddf['Age'], errors='coerce')
```
