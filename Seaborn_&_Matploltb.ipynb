```python
import seaborn as sns
import matplotlib.pyplot as plt

# Built-in datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
diamonds = sns.load_dataset('diamonds')
penguins = sns.load_dataset('penguins')

# Set seaborn theme
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='day', y='total_bill', data=tips, palette='Blues_d', ci=None)
plt.title('Average Total Bill per Day (Tips Dataset)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='species', data=penguins, palette='Set2')
plt.title('Count of Penguin Species')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips, palette='cool', s=100)
plt.title('Tip vs Total Bill (Colored by Time)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set3')
plt.title('Box Plot of Total Bill by Day & Sex')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True, palette='Pastel1')
plt.title('Violin Plot of Total Bill by Day & Sex')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, color='black')
plt.title('Strip Plot of Total Bill by Day')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', palette='Set2')
plt.title('Swarm Plot of Total Bill by Day & Sex')
plt.show()
```

```python
sns.pairplot(iris, hue='species', palette='Set1', height=2.5)
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(tips['total_bill'], shade=True, color='green')
plt.title('KDE Plot of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='time', palette='cool')
plt.suptitle('Joint Plot of Total Bill vs Tip', y=1.02)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='Set1', height=5)
plt.title('Linear Regression of Tip vs Total Bill by Sex')
plt.show()
```

```python
g = sns.FacetGrid(tips, col='time', row='sex', margin_titles=True)
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip', hue='day')
g.add_legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Numeric columns only
iris_numeric = iris.select_dtypes(include='number')

plt.figure(figsize=(8,6))
sns.heatmap(iris_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Iris Dataset - Numeric Columns Only)')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.rugplot(tips['total_bill'], height=0.1, color='red')
plt.title('Rug Plot of Total Bill')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='day', y='total_bill', data=tips, inner=None, palette='Pastel2')
sns.swarmplot(x='day', y='total_bill', data=tips, color='k', alpha=0.6)
plt.title('Violin + Swarm Combo Plot')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hexbin(tips['total_bill'], tips['tip'], gridsize=20, cmap='Blues', edgecolors='grey')
plt.colorbar(label='Counts')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Hexbin / 2D Histogram')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='day', y='total_bill', size='tip', data=tips, hue='sex', palette='Set1', sizes=(20,200))
plt.title('Swarm-Style Plot with Size Scaling by Tip')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='sex', data=tips, palette='Set2')
plt.title('Count by Sex (Tips Dataset)')
plt.show()
```

```python
plt.figure(figsize=(6,6))
tips_day_counts = tips['day'].value_counts()
plt.pie(tips_day_counts, labels=tips_day_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Pie Chart of Tips by Day')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sex_counts = tips['sex'].value_counts()
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title('Pie Chart by Sex')
plt.show()
```
