# Data Visualization using Matplotlib - Tutorial 

## What is Data Visualization?  
Data visualization is the graphical representation of information and data

### Benifits:  
Simplifies complex data  
Identifies patterns, trends, and outliers  
Aids in decision-making

### Real-life Examples:  
1. Health-care: Patient data visualizations to monitor health metrics  
2. Finance: Stock market analysis chart

Matplotib is a comprehensive library for creating static, animated and interactive visualizations in Python

## Choosing right plot:

| **What You Want to Do**                      | **Best Plot Type(s)**             | **What It Shows**                                | **Example Use**                     |
| -------------------------------------------- | --------------------------------- | ------------------------------------------------ | ----------------------------------- |
| **See how values change over time**          | Line plot, Area plot              | Trend or progress over time                      | Training accuracy vs. epochs        |
| **Compare different groups or categories**   | Bar plot, Grouped bar             | Which group has higher or lower values           | Comparing average marks of classes  |
| **See how data is spread out**               | Histogram, Box plot               | Distribution, range, and outliers                | Checking age or salary distribution |
| **Find relationship between two variables**  | Scatter plot                      | Whether one value increases with another         | Height vs. Weight                   |
| **Check proportions or percentages**         | Pie chart, Donut chart            | How much each category contributes to total      | Share of students by department     |
| **Show correlation between features**        | Heatmap                           | Strength of relationships                        | Feature correlation matrix          |
| **Show performance of classification model** | Heatmap (Confusion Matrix)        | Which classes are predicted correctly or wrongly | Model results visualization         |
| **Show importance of features**              | Horizontal Bar plot               | Which features matter most                       | Feature importance from ML model    |
| **Visualize clusters or patterns**           | Scatter plot (after PCA or t-SNE) | How data points group together                   | Clustering visualization            |


```python
import matplotlib.pyplot as plt
```

## 1. Line plot

```python
x = [1,2,3,4,5]
y = [10, 20, 25, 30, 35]

# 1
plt.plot(x, y)

plt.show()
```

#### Customizing plots using [color, linestyle, markers] 

```python
# 2
plt.plot(x, y, color='red')
plt.show()
```

```python
# 3
plt.plot(x, y, color='red', linestyle='--')
plt.show()
```

```python
# 4
plt.plot(x, y, color='red', linestyle='--', marker='o')

plt.show()
```

#### Adding Legends, x-label, y-label and title to plot

```python
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line PLot')
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', label='Line-1')
plt.plot(x, [15, 25, 35, 45, 55], color='blue', linestyle='-', marker='x', label='Line-2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line PLot')
plt.legend()
plt.show()
```

## 2. Bar plot:  
A bar plot (or bar chart) in Matplotlib is used to represent categorical data with rectangular bars — where the length or height of each bar is proportional to the value it represents.

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 17, 35, 29]
```

### a. Vertical Bars

```python
plt.bar(categories, values, color='skyblue')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot Example')

# Show plot
plt.show()
```

### b.HOrizontal bars

```python
plt.barh(categories, values, color='lightgreen')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot')
plt.show()
```

### Customize bar

```python
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'], edgecolor='black', width=0.6)
plt.title('Customized Bar Plot')
plt.show()
```

```python
import numpy as np
# Data
categories = ['A', 'B', 'C', 'D']
men = [20, 34, 30, 35]
women = [25, 32, 34, 20]

x = np.arange(len(categories))  # positions
width = 0.35  # width of bars

plt.bar(x- width/2 , men, width, label='Men', color='lightblue')
plt.bar(x + width/2, women, width, label='Women', color='lightpink')

plt.xlabel('Categories')
plt.ylabel('Scores')
plt.title('Grouped Bar Plot')
plt.xticks(x, categories)
plt.legend()
plt.show()
```

## 3. Histogram:  
A histogram in Matplotlib is used to show the distribution of numerical data — it divides data into bins (intervals) and shows how many values fall into each bin.

```python
# Sample data
data = [12, 15, 13, 17, 19, 21, 22, 25, 25, 29, 30, 31, 33, 33, 35, 37, 40]

# Create histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Simple Histogram')

# Show plot
plt.show()
```

### Multiple Histograms

```python
x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(2, 1, 1000)

plt.hist([x1, x2], bins=20, color=['blue', 'red'], label=['x1', 'x2'], alpha=0.7)
plt.legend()
plt.title('Multiple Histograms')
plt.show()
```

# 4. Scatter plot:  
A scatter plot in Matplotlib is used to display the relationship between two numerical variables — one on the x-axis and one on the y-axis.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 7, 8]

plt.scatter(x, y, color='blue', marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')

plt.show()
```

```python

# Generate data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)              # different colors

plt.scatter(x, y, c=colors, alpha=0.7, cmap='viridis', edgecolor='black')
plt.title('Customized Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.colorbar(label='Color Intensity')  # add color scale
plt.show()
```
